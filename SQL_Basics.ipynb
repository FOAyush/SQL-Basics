#2 2. Explain the purpose of constraints and how they help maintain data integrity in a database. Provide examples of common types of constraints.


###Ans)- Constraints in a database are rules enforced on data columns to ensure accuracy, consistency, and integrity of the data stored. They help prevent invalid data from being entered into the database, ensuring that the information remains reliable and meaningful over time. For instance, a **PRIMARY KEY** constraint ensures that each record in a table is uniquely identifiable, while a **NOT NULL** constraint prevents important fields from being left empty. The **UNIQUE** constraint guarantees that no two rows have the same value in a specific column, which is useful for fields like email or usernames. A **CHECK** constraint enforces rules on the values in a column, such as ensuring an employee’s age is at least 18. Additionally, the **DEFAULT** constraint can automatically assign a value to a column if none is provided, helping maintain consistency. By using these constraints, databases are protected from common errors like duplicate entries, missing data, or logically incorrect values, ultimately maintaining the integrity and quality of the data.


#3.Why would you apply the NOT NULL constraint to a column? Can a primary key contain NULL values? Justify your answer


###Ans)-The **NOT NULL** constraint is applied to a column to ensure that a value is always provided for that field when a new record is inserted. This is crucial for columns that are essential for identifying or describing a record—such as a name, date, or any other required attribute. By enforcing NOT NULL, the database ensures that no record can be created or updated without meaningful data in that column, helping maintain data completeness and reliability.

A **primary key cannot contain NULL values**. This is because the primary key is used to uniquely identify each row in a table. If a primary key were allowed to have NULLs, it would violate this uniqueness rule, since NULL represents an unknown or undefined value, and multiple NULLs would not be distinguishable. Therefore, most relational databases automatically enforce NOT NULL on primary key columns, even if you don’t explicitly specify it. This ensures that every record has a unique, non-null identifier.


#4. Explain the steps and SQL commands used to add or remove constraints on an existing table. Provide an example for both adding and removing a constraint


###ANS)-To modify constraints on an existing table, SQL provides `ALTER TABLE` statements that allow you to add or remove constraints without recreating the table. To **add a constraint**, you use the `ALTER TABLE` command along with `ADD CONSTRAINT`, specifying the constraint type and its details. For example, to add a `UNIQUE` constraint on the `email` column of an `employees` table, you could write:

```sql
ALTER TABLE employees ADD CONSTRAINT unique_email UNIQUE(email);
```

This ensures that no two employees can have the same email address.

To **remove a constraint**, you again use the `ALTER TABLE` command, but with `DROP CONSTRAINT`, followed by the name of the constraint. For instance, to remove the `unique_email` constraint you added earlier, you'd write:

```sql
ALTER TABLE employees DROP CONSTRAINT unique_email;
```

It's important to note that the exact syntax can vary slightly depending on the database system (e.g., MySQL, PostgreSQL, SQL Server), and in some systems, constraints like `NOT NULL` may require altering the column directly rather than using `DROP CONSTRAINT`. These commands are powerful tools for maintaining and evolving the integrity rules of your database as your application’s requirements change.


#5. Explain the consequences of attempting to insert, update, or delete data in a way that violates constraints. Provide an example of an error message that might occur when violating a constraint


###Ans)-When you attempt to **insert**, **update**, or **delete** data in a way that violates constraints, the database will **reject the operation** and return an **error message**. Constraints are enforced to maintain data integrity, so any operation that breaks these rules is blocked to prevent corrupt or invalid data from entering the system.

For example, if you try to insert a record into a table with a **NOT NULL** constraint on a column but leave that column empty, the database will raise an error. Similarly, if you attempt to insert a duplicate value into a column with a **UNIQUE** constraint, or delete a row that is referenced by a **foreign key** in another table, you'll receive constraint violation errors.

###  Example:

Suppose you have a `users` table with a `UNIQUE` constraint on the `email` column. If you try to insert a duplicate email, you might see an error like:

```sql
ERROR: duplicate key value violates unique constraint "users_email_key"
DETAIL: Key (email)=(john@example.com) already exists.
```

These errors prevent the database from accepting data that could lead to inconsistencies, ensuring that all operations respect the defined structure and rules of the schema. Ignoring these constraints would risk introducing duplicate, incomplete, or logically invalid records into your system.

#11 Identify the primary keys and foreign keys in maven movies db. Discuss the differences

Ans)- In a typical Maven movies database, the **primary keys** are unique identifiers for each table that ensure every record can be distinctly identified. For example, in a `movies` table, the primary key might be `movie_id`, uniquely identifying each movie entry. Similarly, in a `actors` table, `actor_id` serves as the primary key. On the other hand, **foreign keys** are fields in one table that reference the primary key of another table to establish relationships between the data. For instance, a `movie_actors` table might contain `movie_id` and `actor_id` as foreign keys linking movies to their actors. The primary key enforces uniqueness and non-nullability for records within its own table, while foreign keys enforce referential integrity by ensuring that the values in one table correspond to existing values in another, preventing orphaned records. In summary, primary keys uniquely identify records within a table, whereas foreign keys create meaningful connections between related tables in the database.

Normalisation & CTE
#1. First Normal Form (1NF)
A table violates 1NF if columns hold multiple values (non-atomic). For example, if actor has a column with a list of movies, convert it to separate rows in a joining table: film_actor.

#2. Second Normal Form (2NF)
2NF requires 1NF, plus all non-key columns must depend on the whole composite key. In a table where some columns depend only on part of the key, split them into new tables so every non-key column depends on the whole key.

#3. Third Normal Form (3NF)
3NF requires 2NF and removal of transitive dependencies. If table payment contains both customer_address and customer_id, move customer_address to a separate customer table referenced by customer_id.

#4. Normalisation Process Example
Pick rental table:

Achieve 1NF: Ensure each column is atomic

Achieve 2NF: Move address details outside if composite keys exist

Further normalise to 3NF if transitive dependencies exist.
